# Linear Regression Exercises

Exercise 8. Use the `Auto` dataset to:

  Perform linear regression using `mpg` as response and `horsepower` as predictor

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
pd.set_option('display.mpl_style', 'default')

auto =  pd.read_csv('/home/nephilim/Documents/Datasets/Auto.csv') #Load the data
XY = zip(list(auto['horsepower']), list(auto['mpg'])) #Make tuples of the variables we need
XY = filter(lambda x: x[0].isdigit(), XY)  #filter anything that is not a number from the data
X, Y = zip(*XY) #unzip the array of tuples
X = np.array([int(x) for x in X])     #in order to be able to use scikit-learn, we must convert them to np.array()
Y = np.array(Y)
fig =  plt.figure(figsize=(15,10))
ax =  fig.add_subplot(1,1,1)
ax.scatter(X,Y, marker='x',c='r')
ax.set_title("mpg vs horsepower")
X = X.reshape(len(X), 1)
model = linear_model.LinearRegression()
model.fit(X, Y)
print "%f0.4*X + %f0.4" % (model.coef_, model.intercept_)
```
